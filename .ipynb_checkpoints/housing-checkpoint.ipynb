### Numerical Feautures

```python
num_features = housing.select_dtypes(include=[np.number])
num_features.columns
```

**Finding Correlation coefficients between numeric features and SalePrice**

```python
correlation = num_features.corr()
print(correlation['SalePrice'].sort_values(ascending = False),'\n')
```

```python
# Correlation Heatmap
f , ax = plt.subplots(figsize = (14,12))

plt.title('Correlation of Numeric Features with Sale Price',y=1,size=16)
sns.heatmap(correlation,square = True,  vmax=0.8)
```

```python
# Correlation of saleprice
# Zoomed Heatmap
k= 11
cols = correlation.nlargest(k,'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(housing[cols].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

From above zoomed heatmap it is observed that GarageCars & GarageArea are closely correlated . Similarly TotalBsmtSF and 1stFlrSF are also closely correlated.

OverallQual', 'GrLivArea' and 'TotalBsmtSF' are strongly correlated with 'SalePrice'.

```python
sns.set()
columns = ['SalePrice','OverallQual','TotalBsmtSF','GrLivArea','GarageArea','FullBath','YearBuilt','YearRemodAdd']
sns.pairplot(housing[columns], height = 2 ,kind ='reg',diag_kind='kde')
plt.show()
```

**Scatter Plot**

Scatter plots between the most correlated variables

```python
fig, ((ax1, ax2), (ax3, ax4),(ax5,ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(14,10))
OverallQual_scatter_plot = pd.concat([housing['SalePrice'],housing['OverallQual']],axis = 1)
sns.regplot(x='OverallQual',y = 'SalePrice',data = OverallQual_scatter_plot,scatter= True, fit_reg=True, ax=ax1)
TotalBsmtSF_scatter_plot = pd.concat([housing['SalePrice'],housing['TotalBsmtSF']],axis = 1)
sns.regplot(x='TotalBsmtSF',y = 'SalePrice',data = TotalBsmtSF_scatter_plot,scatter= True, fit_reg=True, ax=ax2)
GrLivArea_scatter_plot = pd.concat([housing['SalePrice'],housing['GrLivArea']],axis = 1)
sns.regplot(x='GrLivArea',y = 'SalePrice',data = GrLivArea_scatter_plot,scatter= True, fit_reg=True, ax=ax3)
GarageArea_scatter_plot = pd.concat([housing['SalePrice'],housing['GarageArea']],axis = 1)
sns.regplot(x='GarageArea',y = 'SalePrice',data = GarageArea_scatter_plot,scatter= True, fit_reg=True, ax=ax4)
FullBath_scatter_plot = pd.concat([housing['SalePrice'],housing['FullBath']],axis = 1)
sns.regplot(x='FullBath',y = 'SalePrice',data = FullBath_scatter_plot,scatter= True, fit_reg=True, ax=ax5)
YearBuilt_scatter_plot = pd.concat([housing['SalePrice'],housing['YearBuilt']],axis = 1)
sns.regplot(x='YearBuilt',y = 'SalePrice',data = YearBuilt_scatter_plot,scatter= True, fit_reg=True, ax=ax6)
YearRemodAdd_scatter_plot = pd.concat([housing['SalePrice'],housing['YearRemodAdd']],axis = 1)
YearRemodAdd_scatter_plot.plot.scatter('YearRemodAdd','SalePrice')
```

### Related data

All the Garage Related Feature - ['GarageCond', 'GarageQual', 'GarageType' and 'GarageFinish']

```python
fig, axes = plt.subplots(2,2, figsize=(12,10)) 

housing.GarageCond.value_counts().sort_index().plot.bar(ax=axes[0][0]) 
axes[0][0].set_title("GarageCond", fontsize=14)

housing.GarageQual.value_counts().sort_index().plot.bar(ax=axes[0][1]) 
axes[0][1].set_title("GarageQual", fontsize=14)

housing.GarageFinish.value_counts().sort_index().plot.bar(ax=axes[1][0]) 
axes[1][0].set_title("GarageFinish", fontsize=14)

housing.GarageType.value_counts().sort_index().plot.bar(ax=axes[1][1]) 
axes[1][1].set_title("GarageType", fontsize=14)
```

```python
# Replace null values
housing['GarageCond'] = housing['GarageCond'].fillna('None') 
housing['GarageQual'] = housing['GarageQual'].fillna('None') 
housing['GarageType'] = housing['GarageType'].fillna('None') 
housing['GarageFinish'] = housing['GarageFinish'].fillna('None') 
```

### Plotting Dependencies on Sale Price

```python
fig, axis = plt.subplots(2,2, figsize = (12,10))
sns.boxplot(data=housing, x = 'GarageCond', y = 'SalePrice', ax=axis[0,0])
axis[0,0].set_title('GarageCond Vs Sale Price' )

sns.boxplot(data=housing, x = 'GarageQual', y = 'SalePrice', ax=axis[0,1])
axis[0,1].set_title('GarageQual Vs Sale Price' )

sns.boxplot(data=housing, x = 'GarageType', y = 'SalePrice', ax=axis[1,0])
axis[1,0].set_title('GarageType Vs Sale Price' )

sns.boxplot(data=housing, x = 'GarageFinish', y = 'SalePrice', ax=axis[1,1])
axis[1,1].set_title('GarageFinish Vs Sale Price')

plt.tight_layout()
```
